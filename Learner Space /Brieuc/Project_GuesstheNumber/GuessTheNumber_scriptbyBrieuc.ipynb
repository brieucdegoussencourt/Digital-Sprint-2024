```python
# Create a function that generates a random number between 1 and 100

import random

secret_number = random.randrange(0, 100)

#	Create a function that generates a random number between 1 and 100
# Check if the player guess is correct, too high or too low and inform him/her
# Returns message accordingly

def play_round(player_guess, guesses_left, player_name, secret_number):
    if player_guess == secret_number:
        return f"Congratulations {player_name}! You've guessed the correct number in {rounds} rounds!", 0
    elif player_guess < secret_number:
        return "Too low " + player_name + "! Try a higher number.", guesses_left - 1
    else:
        return "Too high " + player_name + "! Try a lower number.", guesses_left - 1

# Ask the player to enter name and number
# Count rounds

guesses_left = 10
player_name = input("What is your name?")
rounds = 0

# Create a loop for managing the game
# Ensure that the number is valid and included in the interval 1-100
# Use the functions created above to generate a number and check if the player guessed correctly
# If the player guesses correctly the loop should stop and display congratulations message
# If the number is not guessed after 10 rounds, a message should appear

while guesses_left > 0:
    try:
        player_guess = int(input(f"Guess the number (between 1 and 100), {guesses_left} guesses left: "))
    except ValueError:
        print("Invalid input. Please enter a whole number between 1 and 100.")
        continue
    if not 1 <= player_guess <= 100:
        print("Invalid input. Please enter a whole number between 1 and 100.")
        continue

    rounds += 1

    result_message, guesses_left = play_round(player_guess, guesses_left, player_name, secret_number)
    print(result_message)
    if "Congratulations" in result_message:
        break
    elif guesses_left == 0:
        print(f"You failed. Try again.")
```
