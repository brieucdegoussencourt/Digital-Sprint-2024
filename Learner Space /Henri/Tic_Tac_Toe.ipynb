```python
current_player = "X"
winner = ""
game_on = True
plateau = [["  1","2","3"],["1", "-", "-", "-"],["2", "-", "-", "-"],["3", "-", "-", "-"]]

player_one = input("Nom du premier joueur: ")
print("Le symbole de", player_one, "sera le X")
player_two = input("Nom du deuxième joueur: ")
print("Le symbole de", player_two, "sera le O")


def afficher_plateau(plateau):
  for i in plateau:
    ligne = ""
    for j in i:
      ligne = ligne + j + " "
    print(ligne)


def placement_joueur():
  col = input("Choisissez le numéro de la colonne dans laquelle vous souhaitez placé votre symbole: ")
  while not col.isdigit() or int(col) > 3 or int(col) < 1:
      col = input("Veuillez introduire un chiffre entre 1 et 3: ")
  row = input("Choisissez le numéro de la ligne dans laquelle vous souhaitez placé votre symbole: ")
  while not row.isdigit() or int(row) > 3 or int(row) < 1:
    col = input("Veuillez introduire un chiffre entre 1 et 3: ")
  print(col,row)
  row = int(row)
  col = int(col)
  if plateau[row][col] != "-":
    print("Case occuppée!")
    placement_joueur()
  else:
    plateau[row][col] = current_player


def switch_player():
  global game_on
  global current_player
  global player_one
  global player_two
  print("\n")
  if game_on == True:
    if current_player == "X":
      current_player = "O"
      print(f"<<< {player_two} (O) >>>")
    else:
      current_player = "X"
      print(f"<<< {player_one} (X) >>>")


def check_hori(plateau):
  global winner
  if plateau[1][1] == plateau[1][2] == plateau[1][3] and plateau[1][1] != "-":
    winner = plateau[1][1]
    return True
  elif plateau[2][1] == plateau[2][2] == plateau[2][3] and plateau[2][1] != "-":
    winner = plateau[2][1]
    return True
  elif plateau[3][1] == plateau[3][2] == plateau[3][3] and plateau[3][1] != "-":
    winner = plateau[3][1]
    return True


def check_verti(plateau):
  global winner
  if plateau[1][1] == plateau[2][1] == plateau[3][1] and plateau[1][1] != "-":
    winner = plateau[1][1]
    return True
  elif plateau[1][2] == plateau[2][2] == plateau[3][2] and plateau[1][2] != "-":
    winner = plateau[1][2]
    return True
  elif plateau[1][3] == plateau[2][3] == plateau[3][3] and plateau[1][3] != "-":
    winner = plateau[1][3]
    return True


def check_diag(plateau):
  global winner
  if plateau[1][1] == plateau[2][2] == plateau[3][3] and plateau[1][1] != "-":
    winner = plateau[1][1]
    return True
  if plateau[3][1] == plateau[2][2] == plateau[1][3] and plateau[3][1] != "-":
    winner = plateau[3][1]
    return True


def check_win(plateau):
  global game_on
  if check_hori(plateau) or check_verti(plateau) or check_diag(plateau):
    if winner == "X":
      print("<<<Bravooo", player_one, ", tu as gagné la partie!!!>>>")
      print(afficher_plateau(plateau))
    else:
      print("<<<Bravooo", player_one, ", tu as gagné la partie!!!>>>")
      print(afficher_plateau(plateau))
    game_on = False


def check_draw(plateau):
  global game_on
  for i in plateau:
    if not any("-" in sublist for sublist in plateau):
      game_on = False
      print("<<<La grille est pleine, MATCH NUL!>>>")
      print(plateau)
      break


def jeu():
  while game_on == True:
    afficher_plateau(plateau)
    placement_joueur()
    check_win(plateau)
    check_draw(plateau)
    switch_player()


jeu()
```
