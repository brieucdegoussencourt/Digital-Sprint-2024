## Built-in Functions
The Python interpreter has a number of functions and types built into it that are always available.

A built-in function is a pre-defined function that is readily available for use without the need to import any additional modules. These functions perform common operations, such as mathematical calculations, string manipulation, or working with data structures like lists and dictionaries. Built-in functions help simplify code, improve code readability, and reduce the need to write repetitive code for frequently-used operations.

We have already used them. Remember the `print()` function which allows you to display text.

```python
print("Hello World")
```

There is also the `input()` method which allows the user to enter text.

```python
age = input("How old are you?")
print("Your age", age)
```

The input value will always be a string. We'll use another native function to check it.

```python
print(type(age))
```

Functions that allow cast : ` str(), int (), float() `, etc...

```python
age = input("How old are you?")
print("string :", age, type(age))

age = int(age)  # Cast from string to integer
print("integer :", age, type(age))

age = float(age)  # Cast from int to float
print("float :", age, type(age))

age = str(age)  # Cast from float to string
print("string :", age, type(age))
```

The `len()` function allows us to know how many elements are in a list or string.

```python
print(len("Hello Word"))

arr = ["one", "two"]
print(len(arr))
```

## Let's discover them in practice

Look at Python built-in functions to solve the following problems:

1. Find the length of the string `"Becode is fun!"` and print the result.
2. Convert the string in a list of characters. Print its length.
3. Convert the list of characters in a `set`. Print its length. Can you explain the difference?
4. Create a list of numbers and print its lowest and highest values.
5. Print the sum of the elements of the list that you have created.
6. Sort the list in descending order and print it.
7. Print the absolute value of `-42`.
8. Create a string that contains a basic sum (like `"4 + 3"`). Find a way to print the result of that sum.
9. Round the decimal number `3.14159` to 2 decimal places, and print the result.
10. Create a string with a number. Convert it to an integer and multiply it by `2`. Print the result.

```python

[1] # Find the length of the string "Becode is fun!" and print the result.
string = "Becode is fun!"
print(len(string))
```

```python
[2] # Convert the string in a list of characters. Print its length.
list(string)
```

```python
[3] # Convert the list of characters in a set. Print its length. Can you explain the difference?
set(string)
```

```python
[4] # Create a list of numbers and print its lowest and highest values.
num_list = (1 , 2 , 3 , 4 , 5 , 6 )
print(min(num_list))
print(max(num_list))
```

```python
[5] # Print the sum of the elements of the list that you have created.
num_list = (1 , 2 , 3 , 4 , 5 , 6 )
print(sum(num_list))
```

```python
[6] # Sort the list in descending order and print it.
num_list = [1 , 2 , 3 , 4 , 5 , 6]
num_list.sort(reverse=True)
print(num_list)
```

```python
[7] # Print the absolute value of -42.
number = abs(-42)
print(number)
```

```python
[8] # Create a string that contains a basic sum (like "4 + 3"). Find a way to print the result of that sum.

calcul = ("4 + 3")
print(eval(calcul))
```

```python
[9] # Round the decimal number 3.14159 to 2 decimal places, and print the result.
pi = 3.14159
round(pi, 2)
print(round(pi, 2))
```

```python
[10] # Create a string with a number. Convert it to an integer and multiply it by 2. Print the result.
num = int(14)
print(num * 2)
```
