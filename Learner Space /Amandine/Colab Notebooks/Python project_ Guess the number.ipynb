```python
import random

# import random: Cette ligne importe le module random, qui est utilisé pour générer des nombres aléatoires.

def generate_random_number():
    # Cette fonction définit une fonction appelée generate_random_number().
    # À l'intérieur de cette fonction, random.randint(1, 100) est utilisé pour générer un nombre aléatoire entre 1 et 100, inclus.
    return random.randint(1, 100)

def play_round(secret_number):
    # Cette fonction définit une fonction appelée play_round() qui prend en paramètre le nombre secret.
    # À l'intérieur de cette fonction, une boucle while True est utilisée pour demander à l'utilisateur de deviner un nombre.
    # Si l'entrée de l'utilisateur est invalide (c'est-à-dire non un nombre ou en dehors de la plage autorisée), un message d'erreur est affiché.
    # Sinon, la fonction retourne le nombre deviné.
    while True:
        try:
            guess = int(input("Guess the number (between 1 and 100): "))
            if guess < 1 or guess > 100:
                print("Please enter a number between 1 and 100.")
                continue
            return guess
        except ValueError:
            print("Invalid input. Please enter a number.")

def manage_game():
    # Cette fonction gère le déroulement du jeu.
    # À l'intérieur de cette fonction, la variable secret_number est initialisée en appelant generate_random_number() pour générer un nombre aléatoire.
    # Ensuite, une boucle for est utilisée pour exécuter chaque round du jeu (10 rounds maximum).
    """Vous débutez à 1, cela signifie que le premier tour est le numéro 1, et si vous terminez à 11, cela signifie que le dernier tour est le numéro 10."""
    # Dans chaque round, la fonction play_round() est appelée pour permettre au joueur de deviner un nombre.
    # Si le nombre deviné par le joueur est égal au nombre secret, la fonction retourne le numéro du round et True.
    # Sinon, le programme informe le joueur si le nombre deviné est supérieur ou inférieur au nombre secret.
    secret_number = generate_random_number()
    for round_num in range(1, 11):
        print(f"Round {round_num}:")
        guess = play_round(secret_number)
        if guess == secret_number:
            return round_num, True
        elif guess < secret_number:
            print("The number is higher.")
        else:
            print("The number is lower.")
    return 10, False

def main():
    # Cette fonction est la fonction principale du programme.
    # Elle commence par demander le nom du joueur.
    # Ensuite, elle appelle la fonction manage_game() pour commencer le jeu.
    # En fonction du résultat retourné par manage_game(), elle affiche un message de félicitations si le joueur a réussi à deviner le nombre, sinon elle affiche un message d'échec.
    player_name = input("Enter your name: ")
    rounds, success = manage_game()
    if success:
        print(f"Congratulations, {player_name}! You guessed the correct number in {rounds} rounds.")
    else:
        print("You failed, try again!")

if __name__ == "__main__":
    # Cette condition vérifie si le programme est exécuté en tant que script principal.
    # Si c'est le cas, la fonction main() est appelée pour démarrer le jeu.
    main()

```

<<Description du projet

Ce projet consiste en la création d'un jeu dans lequel l'ordinateur génère un nombre aléatoire entre 1 et 100, et le joueur doit deviner ce nombre en un maximum de 10 essais. Le jeu fournit des indications sur la position relative de la devinette du joueur par rapport au nombre généré. Si le joueur devine correctement le nombre, il remporte la partie. Sinon, il perd s'il épuise ses 10 tentatives.
Structure du code

Le code est structuré en utilisant des fonctions pour gérer différentes parties du jeu :

    generate_random_number() : Génère un nombre aléatoire entre 1 et 100.
    play_round(secret_number) : Gère une manche de jeu où le joueur devine un nombre.
    manage_game() : Gère le déroulement complet du jeu, y compris l'initialisation, les tours de jeu et la vérification de la victoire ou de la défaite.
    main() : Fonction principale qui demande le nom du joueur et lance le jeu.
