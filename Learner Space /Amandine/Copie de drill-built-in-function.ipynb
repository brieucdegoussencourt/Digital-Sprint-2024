# Drill : built-in functions
Use the native python functions to do these exercises.

### 1. Create a variable `count_alpha` that contains the number of characters contained in the string `"Hello world!"`.

```python
# String
my_string = "Hello world!"

# Number of characters
count_alpha = len(my_string)

# Print result
print("Number of characters:", count_alpha)
```

### 2. Create a variable `count_float` and cast the variable `count_alpha` to float.

```python
# Variable count_alpha containing the number of characters
count_alpha = 12

# Cast count_alpha to float and assign it to count_float
count_float = float(count_alpha)

# Print result
print("Count as float:", count_float)
```

### 3. Round the variable `pi` value to 2 decimal places (0.01) and save it in a variable `round_pi`. You can get `pi` from the `math` library.

```python
import math

# Value of pi from math library
pi = math.pi

# Round pi to 2 decimal places, saved in round_pi
round_pi = round(pi, 2)

# Print result
print("Rounded value of pi:", round_pi)
```

### 4. Create a variable `reversed_text` which contains the character string `"Hello world!"` backwards.
The result must be a `list` object.

```python
# String

original_text = "Hello world!"

# Reverse the string using slicing and save it in reversed_text
reversed_text = list(original_text[::-1])

# Print result
print("Reversed text as list:", reversed_text)
```

### 5. Create a variable `age`  and ask the user to enter his age. Then display it and display its type.

```python
# User age
age = input("Please enter your age: ")

# Display the entered age and its type
print("Entered age:", age)
print("Type of age variable:", type(age))
```

### 6. Create a variable `sorted_num` that contains the sorted `num` list.


```python
#num list
num = [2, 8, 1, 4, 6, 3, 7]

#Sorted numbers in the list
sorted_num = sorted(num)

#print result
print("Sorted numbers in the list:", sorted_num)
```

### 7. Create a variable `sum_of_list` which contains the sum of all the elements in the list `num`.

```python
#num list
num = [2, 8, 1, 4, 6, 3, 7]

#sum of elements in the list
sum_of_list = sum(num)

#print sum
print("Sum of the elements in the list:", sum_of_list)
```

### 8. Create a variable `min_value` that contains a minimum value of list `num`.

```python
#num list
num = [2, 8, 1, 4, 6, 3, 7]

# min value in the list
min_value = min(num)

#print result
print("Minimum value in the list:", min_value)

```

### 9. Create a variable `max_value` that contains the maximum value of list `num`.

```python
#num list
num = [2, 8, 1, 4, 6, 3, 7]

# Max value in the list
max_value = max(num)

#print result
print("Maximum value in the list:", max_value)
```

### 10. Find a function that will interpret the string of the variable `calc`
Save the result in a variable named `string_interpret`.

```python
# string
calc = "1 + 2"

# Evaluate the expression using eval()
string_interpret = eval(calc)

# Display the result
print("Result of the calculation:", string_interpret)
```
