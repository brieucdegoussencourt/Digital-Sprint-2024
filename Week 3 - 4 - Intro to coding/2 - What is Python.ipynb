## Python's philosophy

Python's design philosophy prioritizes simplicity, readability, and ease of use, as outlined in the Zen of Python, a collection of guiding principles written by Tim Peters. Key principles include:

1. Readability counts: Python emphasizes clear and easily understandable code, using indentation to define code blocks and avoiding complex syntax.
2. Less is more: Python values simplicity and conciseness, preferring a obvious way to achieve a task rather than multiple and complex ways.
3. Explicit is better than implicit: Python encourages code that is easy to understand and avoids hidden behaviors or assumptions.
4. Practicality over purity: While Python strives for simplicity and elegance, it also acknowledges the importance of practical solutions to real-world problems.

### What is executed and what is not?

Basically, everything that you write in your code will be executed by Python.

If you want to avoid that a piece of code is executed you can use **comments**. To do so you will put a hashtag (`#`) before the part of code you don't want to execute. All characters after the `#` and up to the end of the physical line are part of the comment and the Python interpreter ignores them.

```python
# My first comment
name = "BeCode"  # Name of school
```

Following triple-quoted string is also ignored by Python interpreter and can be used as a multiline comments:

```python
"""
This is a multiline
comment.
"""
```

Did you notice that the color of the words in your code changes based on what you type? This process is called **syntax highlighting**. Each Python element is associated to a certain color and it helps programmers to quickly distinguish between different parts of the code. This is not implemented in Python itself but this is used in most IDEs (integrated development environments). An IDE is a software application that provides a comprehensive set of tools and features to assist programmers in writing, editing, debugging, and running code efficiently. Google Colab is considered as a web-based IDE dedicated to Python only.
